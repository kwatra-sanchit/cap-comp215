Notice that the `miss_distance` field contains the distance (in various units) by which the NEO missed an "orbiting body".

## Exercise 2

In the code cell below, write a python function that takes a list of "close approach data" as a parameter,
and returns a 2-tuple with the (date, miss km) of the closest approach to Earth in the list (where "miss km" is the miss distance in km).

Hints:
* notice the input is a list of dictionaries.  Each dictionary has a 'close_approach_date", "orbiting_body", and 'miss_distance' field.
* we are only interested in the closest approach to "Earth"
* use a loop if that is easier to understand - we will look at more compact algorithms to solve this problem in class.

Add at least one unit test to check your work - note the test data only needs dictionaries with the fields your function actually uses.


```python
def get_closest_earth_approach(close_approach_data_list):
    """
    Finds the closest approach to Earth from a list of close approach data.
    """
    # 1. Filter for Earth approaches and extract (date, distance)
    earth_approaches = []
    for approach in close_approach_data_list:
        if approach['orbiting_body'] == 'Earth':
            date = approach['close_approach_date']
            distance = float(approach['miss_distance']['kilometers'])
            earth_approaches.append((date, distance))

    # 2. If no Earth approaches exist, return None
    if not earth_approaches:
        return None

    # 3. Find the minimum distance using a standard loop
    closest_approach = earth_approaches[0] # Start by assuming the first one is the closest

    for current_approach in earth_approaches:
        # Compare current distance with the smallest distance found so far
        if current_approach[1] < closest_approach[1]:
            closest_approach = current_approach

    return closest_approach

test_data_1 = [
        {'orbiting_body': 'Earth', 'close_approach_date': '2026-01-01', 'miss_distance': {'kilometers': '500000'}},
        {'orbiting_body': 'Earth', 'close_approach_date': '2026-01-05', 'miss_distance': {'kilometers': '100000'}}, # Closest
        {'orbiting_body': 'Earth', 'close_approach_date': '2026-01-10', 'miss_distance': {'kilometers': '300000'}}
    ]
expected_1 = ('2026-01-05', 100000.0)
assert get_closest_earth_approach(test_data_1) == expected_1
print("Test Case 1 Passed: Correctly identified closest approach.")
```

In the code cell below, write a complete program that:
 1. fetches the list of NEO's for this week.
 2. for each NEO, fetch it's complete orbital data and determine its closest approach to Earth
 3. identify which NEO from this week's data makes the closet approach to earth
 4. print a nice message with information about the NEO, when it will approach the Earth, and how close it will come.

Hints:

```python
from datetime import timedelta, date

# 1. Get dates for the next 7 days
start_date_obj = date.today()
end_date_obj = start_date_obj + timedelta(days=7)

# 2. Fetch the weekly feed
feed_params = {
    'start_date': start_date_obj.strftime('%Y-%m-%d'),
    'end_date': end_date_obj.strftime('%Y-%m-%d'),
    'api_key': API_KEY # Corrected variable name
}
response = requests.get(NASA_NEO_FEED_API_URL, params=feed_params)
response.raise_for_status()
feed_data = response.json()

# 3. Variables to track the "winner" (closest asteroid)
overall_min_dist = float('inf') # Start with infinity so any distance is smaller
closest_asteroid_msg = ""

# 4. Iterate through each day and each asteroid in that day
for day in feed_data['near_earth_objects']:
    for neo in feed_data['near_earth_objects'][day]:

        # Fetch full data for this specific asteroid to get all approach history
        neo_id = neo['id']
        # Assuming NASA_NEO_API_URL is defined from previous cells
        lookup_url = NASA_NEO_API_URL.format(id=neo_id)
        neo_res = requests.get(lookup_url, params={'api_key': API_KEY}) # Corrected variable name
        neo_res.raise_for_status()
        full_data = neo_res.json()

        # Use Exercise 2 function (get_closest_earth_approach)
        result = get_closest_earth_approach(full_data['close_approach_data'])

        # 5. Check if this asteroid is the closest one seen so far
        if result is not None:
            current_dist = result[1]
            if current_dist < overall_min_dist:
                overall_min_dist = current_dist
                name = full_data['name']
                approach_date = result[0]
                closest_asteroid_msg = f"Asteroid {name} will be closest on {approach_date} at {current_dist:,.2f} km."

# 6. Print the result
print(closest_asteroid_msg)
```
